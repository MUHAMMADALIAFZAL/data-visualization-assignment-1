```python
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline
sns.set(color_codes=True)
```

```python
df = pd.read_csv("data.csv")
# To display the top 5 rows
df.head(5)
```

```python
df.Make.value_counts().nlargest(15).plot(kind='bar', figsize=(10,5))
plt.title("Price of cars by make")
plt.ylabel('price of cars')
plt.xlabel('Make');
```

```python
df.Make.value_counts().nlargest(15).plot(kind='line', figsize=(10,5))
df.Model.value_counts().nlargest(40).plot(kind='line', figsize=(10,5))

plt.title("Line Graph")

plt.xlabel('Make');
```

```python
df.Make.value_counts().nlargest(40).plot(kind='pie', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make');
```

```python
df.Year.value_counts().plot(kind='pie')
plt.show()
```
